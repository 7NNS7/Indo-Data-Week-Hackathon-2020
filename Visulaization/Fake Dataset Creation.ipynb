```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def giveHeatMap(l):
    plt.figure(dpi = 100)
    return sns.heatmap(l,annot = True, xticklabels = False,yticklabels = False)
```

```python
l = [[5,4,4,3,1],
     [4,4,2,1.5,1],
     [4,3,2,1.5,1],
     [2,1.5,1,1,1]]
```

```python
giveHeatMap(l)
```

```python
import random
value = "sensordata_"
d = {}
for i in range(25):
    random.shuffle
    d[value+str(i)] = random.triangular(i,i+10)
    
```

```python
l = []
for i in range(5):
    x = []
    for j in range(5):
        temp = list(d.values())
        x.append(temp[i+j])
    l.append(x)
```

```python
giveHeatMap(l)
```
